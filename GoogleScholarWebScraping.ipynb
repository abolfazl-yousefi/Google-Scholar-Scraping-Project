```python
import pandas as pd
from scholarly import scholarly, ProxyGenerator
import csv

pg = ProxyGenerator()
success = pg.FreeProxies()
scholarly.use_proxy(pg)

keyword = "ocean engineering"
search_query = scholarly.search_keyword(keyword)
results = []
counter = 0
max_authors = 10

fieldnames = ['name', 'affiliation', 'email_domain', 'interests', 'citedby', 'scholar_id']
with open(f'author_info.csv', mode='w') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    for author in search_query:
        author_info = {
            'name': author['name'],
            'affiliation': author['affiliation'],
            'email_domain': author['email_domain'],
            'interests': author['interests'],
            'citedby': author['citedby'],
            'scholar_id': author['scholar_id']
        }
        writer.writerow(author_info)
        results.append(author_info)
        counter += 1
        if counter == max_authors:
            break
    df = pd.DataFrame(results)
    unique_interests = tuple(set(sum(df["interests"], [])))

    for interest in unique_interests:
        search_query = scholarly.search_keyword(interest)
        counter = 0
        for author in search_query:
            author_info = {
                'name': author['name'],
                'affiliation': author['affiliation'],
                'email_domain': author['email_domain'],
                'interests': author['interests'],
                'citedby': author['citedby'],
                'scholar_id': author['scholar_id']
            }
            writer.writerow(author_info)
            results.append(author_info)
            counter += 1
            if counter == max_authors:
                break

    if results:
        df_temp = pd.DataFrame(results)
        df_temp.to_excel(f"authors.xlsx", index=False)

    print(df_temp)
```
